** Agrupamento de Dados e Plotagem de Gráficos**

1- Número de trabalhadores por departamento

```python
df_grupo1 = df.groupby('Department')[['EmployeeCount']].count()
df_grupo1.sort_values('Department')
```

Gráfico representativo

```python
df_grupo1.plot(kind = 'bar')
```

Plote um gráfico de linha (line) que exiba a quantidade de
funcionários para cada nível de escolaridade

```python
df_grupo2 = df.groupby('Education')[['EmployeeCount']].count()
df_grupo2.sort_values('Education')
```

```python
df_grupo2.plot(kind = 'line', marker = 'v')
```

Plote um gráfico de torta/pizza (pie) que exiba a quantidade de
funcionários que tem ou não algum atrito na empresa

```python
df_grupo3 = df.groupby('Attrition')[['EmployeeCount']].count()
df_grupo3.sort_values('Attrition')
```

```python
df_grupo3.plot(kind = 'pie', subplots = True)
```
