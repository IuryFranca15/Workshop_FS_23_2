```python
valor_minimo = df['math score'].min()
valor_maximo = df['math score'].max()
media = df['math score'].mean()

print('valor minimo é: ' ,valor_minimo)
print('valor maximo é: ' ,valor_maximo)
print('media é: ', media)
```
